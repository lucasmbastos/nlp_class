Let a be our previous assigned variable, we will plot the following functions:

* $a$
* $x = a^2$
* $y = 2a$
* $z = log_2(a)$

```python
x = 2*a
y = a ** 2
z = np.log2(a)
```

```python
plt.figure(figsize=(10,5))

plt.plot(a, a)
plt.plot(a, x)
plt.plot(a, y)
plt.plot(a, z)

plt.legend(["a","2*a", "a**2", "log2(a)"])

plt.show()
```
